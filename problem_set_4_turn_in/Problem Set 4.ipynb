```python
# problem 10 
def cleanSample(self):
    LV = []
    RV = []
    for i in self["Sample"]:
        LV.append(i[-2:])
        RV.append(i[:-2])
    for i in range(len(RV)):
        if RV[i] == '':
            RV[i] = 0
        else:
            RV[i] = int(RV[i])
    self["Sample Type"] = LV
    self["Sample Size"] = RV
    self = self.replace('', None)
    return self
```

```python
# problem 12 
def computePollWeight(self, pollName):
    return sum(self[self['Poll'] == pollName]["Sample Size"]) / sum(self["Sample Size"])
```

```python
# problem 13 
def weightedStatsPerCandidate(candidate, self):
    weighted = []
    for poll in self["Poll"].unique():
        x = sum(self[self["Poll"] == poll][candidate])
        y = computePollWeight(self, poll)
        weighted.append(x*y)
    return sum(weighted)/len(weighted)
```

```python
#problem 19 
def getConfidenceInterval(column):
    Array = 1.0 * np.array(column)
    standardError = scipy.stats.sem(Array)
    n = len(column)
    return standardError * scipy.stats.t.ppf((1+.95)/2.0,n-1)
```

```python
getConfidenceInterval(df["Biden"])
getConfidenceInterval(df["Sanders"])
```
